```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn import preprocessing
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier, RadiusNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
df = pd.read_csv('Train_v2.csv')
```

```python
df_encode = pd.read_csv('Train_v2.csv')

le = LabelEncoder()

df_encode['bank_account'] = le.fit_transform(df_encode['bank_account'])
df_encode['location_type'] = le.fit_transform(df_encode['location_type'])
df_encode['cellphone_access'] = le.fit_transform(df_encode['cellphone_access'])
df_encode['gender_of_respondent'] = le.fit_transform(df_encode['gender_of_respondent'])
df_encode['relationship_with_head'] = le.fit_transform(df_encode['relationship_with_head'])
df_encode['education_level'] = le.fit_transform(df_encode['education_level'])
df_encode['job_type'] = le.fit_transform(df_encode['job_type'])

df_encode.drop(['country', 'year', 'uniqueid', 'marital_status'], 1, inplace=True)


df_encode['age_of_respondent'] = df_encode['age_of_respondent'] / max(df_encode['age_of_respondent'].tolist())

df_encode['job_type'] = df_encode['job_type'] / max(df_encode['job_type'].tolist())

df_encode['education_level'] = df_encode['education_level'] / max(df_encode['education_level'].tolist())

df_encode['relationship_with_head'] = df_encode['relationship_with_head'] / max(df_encode['relationship_with_head'].tolist())

df_encode['household_size'] = df_encode['household_size'] / max(df_encode['household_size'].tolist())


train_data = df_encode.values.tolist()

encoded_targets = [i[0] for i in train_data]

df_encode.drop(['bank_account'], 1, inplace=True)

encoded_train_data = df_encode.values.tolist()

encoded_train_data = np.array(encoded_train_data)

encoded_targets = np.array(encoded_targets)

encoded_targets.shape
```

```python
Xe_train, Xe_test, ye_train, ye_test = train_test_split(encoded_train_data, encoded_targets, test_size=0.2, random_state=42)
```

```python
kne_model = KNeighborsClassifier(n_neighbors=17)

kne_model.fit(Xe_train, ye_train)

kne_predicted= kne_model.predict(Xe_test)

print('KNN error rate is: ', 1 - accuracy_score(ye_test, kne_predicted))
```

```python
lre_model = LogisticRegression()

lre_model.fit(Xe_train, ye_train)

lre_predicted= lre_model.predict(Xe_test)

print('Logistic regression error rate is: ', 1 - accuracy_score(ye_test, lre_predicted))
```

```python
rne_model = RadiusNeighborsClassifier(radius=15)

rne_model.fit(Xe_train, ye_train)

rne_predicted= rne_model.predict(Xe_test)

print('radius neighbor error rate is: ', 1 - accuracy_score(ye_test, rne_predicted))
```
